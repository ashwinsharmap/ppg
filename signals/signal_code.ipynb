```python
import pandas as pd
df=pd.read_csv('bidmc_01_Signals.csv')
```

```python
import matplotlib.pyplot as plt
df
```

```python
plt.plot(df.iloc[0:125,0], df.iloc[0:125,2])
plt.grid()
```

```python
df.columns
```

```python
maxi = []
mini = []
for i in range(0,480):
    maxi.append(max(df.iloc[125*i:125*(i+1),2]))
    mini.append(min(df.iloc[125*i:125*(i+1),2]) )               
```

```python
plt.scatter([i for i in range (len(maxi))], maxi)
```

```python
plt.scatter([i for i in range (len(mini))], mini)
```

```python
plt.scatter(maxi,mini)
```

### From the above plot we can observe that most of maximum and minimum points of Pleth wrt time are scattered in the region 0.55 to 0.65

```python
plt.plot(df.iloc[0:125,0], df.iloc[0:125,1])
plt.grid()
```

```python
maxi_resp = []
mini_resp = []
for i in range(0,480):
    maxi_resp.append(max(df.iloc[125*i:125*(i+1),1]))
    mini_resp.append(min(df.iloc[125*i:125*(i+1),1]) )                     
```

```python
plt.scatter([i for i in range (len(maxi_resp))], maxi_resp)
```

```python
plt.scatter([i for i in range (len(mini_resp))], mini_resp)
```

```python
plt.scatter(maxi_resp,mini_resp)
```

### From the above plot we can observe that most of maximum and minimum points of Resp wrt time are scattered in the region 1.0

```python
diff_crest_pleth = []
for i in range(1,len(maxi)):
    diff_crest_pleth.append(abs(maxi[i]-maxi[i-1]))
    
diff_trough_pleth = []
for i in range(1,len(maxi)):
    diff_trough_pleth.append(abs(mini[i]-mini[i-1]))
```

```python

plt.scatter([i for i in range(len(diff_crest_pleth))],diff_crest_pleth)
d = {}
for ele in diff_crest_pleth:
    if ele in d:
        d[ele]=1
    else:
        d[ele]=1
temp = []
for ele in d:
    temp.append(ele)
temp.sort()
temp[len(temp)//2]
```

### From the above scatterplot we can observe that the values of difference between adjacent crests are scattered wrt sample in the range 0 to 0.04

```python
plt.scatter([i for i in range(len(diff_trough_pleth))],diff_trough_pleth)
```

### From the above scatterplot we can observe that the values of difference between adjacent troughs are scattered wrt sample in the range 0 to 0.025

```python
plt.scatter(diff_crest_pleth,diff_trough_pleth)
```

### From the above scatterplot we can observe that the values of difference between adjacent crests are scattered wrt values of difference between adjacent troughs in the range 0.025 to 0.04
